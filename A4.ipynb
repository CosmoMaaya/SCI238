```python
import numpy as np
import matplotlib.pyplot as plt

R = np.array([5, 10, 15, 20, 25])
M_R = np.array([9.25e40, 2.96e41, 4.73e41, 6.21e41, 8.31e41])
plt.figure(figsize=(14,10))
y_output, = plt.loglog(R, M_R, 'o')
# y_output = plt.scatter(R, M_R, label="He Line")
plt.grid(True, which="both")
first_legend = plt.legend(handles=[y_output], loc="upper left")
ax = plt.gca().add_artist(first_legend)

# plt.legend(handles=points, loc="lower right")
m, b = np.polyfit(np.log(R), np.log(M_R), 1)
X = np.logspace(0, 2)
# plt.loglog(x, y, 'o')
plt.loglog(X, np.exp(b) * X ** m, '--')

plt.xlabel("Ratio M/R")
plt.ylabel("Temperature in Kelvin")
plt.title("Temperature Limit for Retention of Atmosphere Gases")
plt.show()
```
